```python
from os.path import dirname, abspath,join
from os import makedirs
import sys
import numpy as np
import pandas as pd
import re
import json
from omegaconf import OmegaConf
import os
import glob
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import seaborn as sns

root_path = dirname(abspath("./"))
import sys
sys.path.append(root_path)
from proT.evaluation.predict import *
from proT.evaluation.metrics import *
from proT.evaluation.eval_sweeps import eval_sweeps

# plotting standard settings
plt.rcParams['figure.dpi'] = 100 #360 # standard is 360 but for set to 100 for practical visualization on the notebook
plt.rcParams['axes.labelsize'] = 18
plt.rcParams['axes.titlesize'] = 18
plt.rcParams['xtick.labelsize'] = 16
plt.rcParams['ytick.labelsize'] = 16
plt.rcParams['legend.fontsize'] = 14
plt.rcParams['figure.figsize'] = (6, 6)
plt.rcParams['lines.linewidth'] = 1.5
plt.rcParams['axes.prop_cycle'] = plt.cycler(color=sns.color_palette('colorblind'))
plt.rcParams['lines.linewidth'] = 2.0
```

```python
datadir_path = "../data/input"
experiment_dir = "../experiments/miss_vals/euler/"  # or whatever directory you want to scan
df_tot = eval_sweeps(
    filepath=experiment_dir,
    datadir_path=datadir_path
)
```

```python
df_tot["p_blank"] = df_tot['model_folder'].apply(lambda x: re.search(r'_([\d]+[.]?[\d]*)$', x).group(1) if re.search(r'_([\d]+[.]?[\d]*)$', x) else None)

options = ["_GRU_", "_proT_", "_proTNoNAIM_"]  # example options
model_pattern = '|'.join(map(re.escape, options))
df_tot['model'] = df_tot['level_0'].apply(lambda x: next((opt for opt in options if opt in x), None))
df_tot["model"] = df_tot["model"].map({
    "_GRU_":"GRU",
    "_proT_":"proT",
    "_proTNoNAIM_":"proT (no NAIM)"})
```

```python
fig = plt.figure()
ax = sns.boxplot(df_tot, x="p_blank", y="MAE", hue="model")
ax.set_yscale("log")
ax.set_xlabel(r"$p_\text{blank}$")
ax.set_ylabel("Test MAE")

ax.legend(loc='upper center',
          bbox_to_anchor=(0.5, 1.1),
          ncol=3,  # One column per legend entry
          frameon=False)

# export
filename = f"missing_vals.pdf"
save_dir = "./figures"
if save_dir is not None:
    makedirs(save_dir, exist_ok=True)
    out_path = join(save_dir, filename)
    fig.savefig(out_path, format="pdf", bbox_inches="tight")
        
plt.show()
```

```python
experiment_dir = "../experiments/ds_size/ds_size_proT_ishigami_sum_120ds_53553862"  # or whatever directory you want to scan
df_120_sum = eval_sweeps(
    filepath=experiment_dir,
    datadir_path=datadir_path
)
```

```python
df_tot['Fusion'] = df_tot['level_0'].apply(lambda x: re.search(r'(sum|cat)', x).group(1) if re.search(r'(sum|cat)', x) else None)
df_tot["Epochs"] = df_tot['model_folder'].apply(lambda x: re.search(r'_(\d+)$', x).group(1) if re.search(r'_(\d+)$', x) else None).astype(int)
df_tot["DS Size"] = df_tot['model_folder'].apply(lambda x: re.search(r'_(\d+)_', x).group(1) if re.search(r'_(\d+)$', x) else None)

df_120_sum["Fusion"] = "sum"
df_120_sum["Epochs"] = df_120_sum['model_folder'].apply(lambda x: re.search(r'_(\d+)$', x).group(1) if re.search(r'_(\d+)$', x) else None).astype(int)
df_120_sum["DS Size"] = 120

feat = ["Epochs", "DS Size", "MAE"]
```

```python
df_plot = pd.concat([df_tot[feat], df_120_sum[feat]], axis=0)
df_plot
```

```python
fig, ax = plt.subplots()

sns.boxplot(df_plot, x="DS Size", y="MAE", hue="Epochs", ax=ax)
#ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
ax.set_xlabel(r"Train Split Size")
ax.set_ylabel(r"Test MAE")
ax.set_yscale("log")
#ax.legend(loc='upper center',
#          bbox_to_anchor=(0.5, 1.18),
#          ncol=3,  # One column per legend entry
#          frameon=False, title="Epochs", title_fontsize=16
#          )

# export
filename = f"ds_size_epochs_mae.pdf"
save_dir = "./figures"
if save_dir is not None:
    makedirs(save_dir, exist_ok=True)
    out_path = join(save_dir, filename)
    fig.savefig(out_path, format="pdf", bbox_inches="tight")
        
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from matplotlib.colors import LogNorm
from os import makedirs
from os.path import join

fig, ax = plt.subplots()

# Extract unique epochs and map them to a log-scaled colormap
unique_epochs = sorted(df_plot["Epochs"].unique())
norm = LogNorm(vmin=min(unique_epochs), vmax=max(unique_epochs))
cmap = plt.cm.viridis
colors = {epoch: cmap(norm(epoch)) for epoch in unique_epochs}

# Create a color palette for seaborn
palette = [colors[epoch] for epoch in unique_epochs]

# Plot the boxplot with the custom palette
sns.boxplot(
    df_plot,
    x="DS Size",
    y="MAE",
    hue="Epochs",
    ax=ax,
    palette=palette,
    hue_order=unique_epochs,
    legend=False
)

ax.set_xlabel(r"Train Split Size")
ax.set_ylabel(r"Test MAE")
ax.set_yscale("log")

# Add a colorbar for the hue
sm = plt.cm.ScalarMappable(norm=norm, cmap=cmap)
sm.set_array([])
cbar = fig.colorbar(sm, ax=ax)
cbar.set_label("Epochs")

# export
filename = f"ds_size_epochs_mae.pdf"
save_dir = "./figures"
if save_dir is not None:
    makedirs(save_dir, exist_ok=True)
    out_path = join(save_dir, filename)
    fig.savefig(out_path, format="pdf", bbox_inches="tight")

plt.show()
```
